```python
import pandas as pd
import numpy as np

data=pd.read_csv('tweets.csv')

data = data.set_index('id')

for i in range(len(data)-1):
    print('Tweet ' + str(i) + ' out of ' + str(len(data)-1))
    print('')
    point = pd.DataFrame(data.iloc[i]).T.copy()
    tweet=point.text.values[0]
    if point.Relevant.values>0:
        continue
    print(tweet)
    print('relevant or irrelevant (1 or 0)')
    rel=int(input())
    if rel==0:
        point.Relevant=0
        data.iloc[i:i+1,:]=point
        data.to_csv('tweets.csv')
        continue
    point.Relevant=1
    print('')
    print('Now you will get a series of classifications. Choose only one of them to match the tweet.')
    print('')
    print('The classifications are: Violent, Incites action, Threat, Anger, Upset, Neutral/political')
    print('')
    print('Below again is the tweet')
    print('')
    print(tweet)
    fin=0
    while(fin==0):
        print('')
        print('Violent (1 or 0)')
        vio=int(input())
        if vio==1:
            point.Violent=1
            fin=1
            continue
        print('Incites action')
        inc=int(input())
        if inc==1:
            point.Incites=1
            fin=1
            continue
        print('Threat')
        thr=int(input())
        if thr==1:
            point.Threat=1
            fin=1
            continue
        print('Anger')
        ang=int(input())
        if ang==1:
            point.Anger=1
            fin=1
            continue
        print('Upset')
        ups=int(input())
        if ups==1:
            point.Upset=1
            fin=1
            continue
        print('Neutral/political')
        nue=int(input())
        if nue==1:
            point.Neutral=1
            fin=1
            continue
        if fin==0:
            print("didnt select any categories, try again")
    data.iloc[i:i+1,:]=point
    
    data.to_csv('tweets.csv')
    print('')
    print('Next Tweet!!!')
    print('========')
    print('')

data=data.iloc[:-1,:]
data.to_csv('tweets.csv')
    
    
```
